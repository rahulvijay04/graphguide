## Idea 2: Common review descriptors

Places with many reviews have aggregations of common descriptors of events

i.e.: see https://www.google.com/maps/place/The+Tap+Haus/data=!4m2!3m1!1s0x80857c2f166365c5:0x15898b0888927c2f!10m1!1e1 - under reviews, there's pool, games, german beer, atmosphere, connect 4, crowd, college, etc.

Divit built a nice selenium code to get these features.

```python
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
from tqdm import tqdm

InteractiveShell.ast_node_interactivity = "all"

berkeley_local = pd.read_csv("comprehensive_berkeley.csv")
pd.set_option('display.max_rows', 4000)
```

```python
# Idea 2: get review descriptor aggregations
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()

def get_keywords(url, attempts):

    try:
        driver.get(url)
    except:
        return get_keywords(url, attempts + 1)
    
    kw = {}
    all_buttons = driver.find_elements(By.CSS_SELECTOR, "button")
    
    for button in all_buttons:
        aria_label = button.get_attribute("aria-label")
        if aria_label and ", mentioned in" in aria_label:
            cat_label = aria_label.split(",")[0]
            ct_label = int(aria_label.split(",")[1].replace("mentioned in ", "").replace(" reviews", ""))
            kw[cat_label] = ct_label
            
    ###### REPEAT
    if len(kw) == 0: 
        if attempts != 5: # If single failure, so far has always had it on second so should be good
            return get_keywords(url, attempts + 1)
    ######
    
    return kw

# get_keywords("https://www.google.com/maps/place/Redwood+Grove+Amphitheatre/data=!4m2!3m1!1s0x80857d80a426afa3:0xae90a289227855b8!10m1!1e1", 0)
```

```python
# remove doubles and unpopulars
berkeley_short = berkeley_local.drop_duplicates().drop("open_hours", axis=1)
berkeley_short = berkeley_short[berkeley_short.review_count >= 100]

# remove unnecessary columns & adjust columns
berkeley_short.drop(['popular_times','website', 'phone', 'plus_code', 'reviews_per_rating', 'cid', 'descriptions',
                    'reviews_link', 'thumbnail', 'timezone', 'data_id', 'images', 'reservations', 'order_online', 'menu', 'owner',
                    'complete_address', 'about', 'user_reviews', 'emails', 'status'], axis=1, inplace=True)
berkeley_short['price_range'] = berkeley_short['price_range'].fillna(0).replace("$", 1).replace("$$", 2).replace("$$$", 3).replace("$$$$", 4)
berkeley_short['features'] = [{}] * berkeley_short.shape[0]
new_column_order = ['title', 'features', 'review_rating', 'review_count', 'price_range', 'category', 'latitude', 'longitude', 'address', 'link']
berkeley_short = berkeley_short[new_column_order]

# rename columns
berkeley_short.rename(columns={'review_rating':'quality', 'price_range':'price', 'review_count': 'popularity'})

print("Count BEFORE:", len(berkeley_local.index))
print("Count AFTER:", len(berkeley_short.index))
```

```python
# Review activities
berkeley_short.head()
```

```python
# Iterate feature extraction
failed = []

for index, row in tqdm(berkeley_short.iterrows()):
    features = get_keywords(row['link'], 0)
    try:
        berkeley_short.loc[berkeley_short.index == index, 'features'] = [features]
    except:
        print("failed!")
        failed.append(row['title'])
    break
print(failed)
```
