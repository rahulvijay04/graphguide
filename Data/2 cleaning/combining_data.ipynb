## Full data together

```python
import pandas as pd
import numpy as np
```

```python
events = pd.read_csv("events_with_locations.csv")
locations = pd.read_csv("activities_with_times.csv")
```

```python
new_columns = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']

for column in new_columns:
    events[column] = np.nan

events["category"] = "Event"
```

```python
events.head()
```

```python
result = pd.concat([events, locations], ignore_index=True)
```

```python
result
```

```python
result = result.drop(columns=["Unnamed: 0"])
```

```python
result.to_csv("events_and_locations.csv")
```
